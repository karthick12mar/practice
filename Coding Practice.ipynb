```python
#test cases
import random
def testCases(total):
    samples = [[],[1],[1,2,3],[1,1,1,1,1],[19, 17, 16, 15, 14, 12, 10, 8, 6, 6, 5, 4, 3, 2, 1]]
    for i in range(total - len(samples)):
        randomlist = []
        listLength = random.randint(5,30)
        for i in range(0,listLength):
            n = random.randint(1,20)
            randomlist.append(n)
        #print(randomlist)
        samples.append(randomlist)
    return samples
```

```python
#function for selection sort

def selectionSort(unSortedList):
    
    if not unSortedList or len(unSortedList) <= 1:
        return unSortedList
    totallen =  len(unSortedList)
    count = 0
    for i in range(totallen):
        pos = i
        for j in range(i,totallen):
            if unSortedList[pos] > unSortedList[j]:
                pos = j
            count = count + 1
        temp = unSortedList[i]
        unSortedList[i] = unSortedList[pos]
        unSortedList[pos] = temp
    print("list len->",len(unSortedList),"\niterations-->",count)
    
    return unSortedList
 
#testcases
totalTests = 10
for test in testCases(totalTests):
    print("input--> ",test,"\noutput--> ",selectionSort(test.copy()),"\n")
    
```

```python
#function for bubble sort


def bubbleSort(unSortedList):
    if not unSortedList or len(unSortedList) <= 1:
        return unSortedList
    totallen = len(unSortedList)
    count = 0
    for i in range(totallen):
        for j in range(totallen-1-i):
            if unSortedList[j] > unSortedList[j+1]:
                temp = unSortedList[j+1]
                unSortedList[j+1] = unSortedList[j]
                unSortedList[j] = temp
            count = count + 1
    print("list len->",len(unSortedList),"\niterations-->",count)
    
    return unSortedList
          
#testcases
totalTests = 10
for test in testCases(totalTests):
    print("input--> ",test,"\noutput--> ",bubbleSort(test.copy()),"\n")
```

```python
#function for Merge sort
def mergeSort(unSortedList):
    a = unSortedList
    if not unSortedList or len(unSortedList) <= 1:
        return unSortedList

    mid = len(a) // 2
    left = unSortedList[:mid]
    right = unSortedList[mid:]

    mergeSort(left)
    mergeSort(right)

    i = j = k = 0
    while(i < len(left) and j < len(right)):
        if left[i] < right[j]:
            unSortedList[k] = left[i]
            i = i + 1
        else:
            unSortedList[k] = right[j]
            j = j + 1
        k = k + 1
        
    while(i<len(left)):
        unSortedList[k] = left[i]
        i = i + 1
        k = k + 1
        
    while(j<len(right)):
        unSortedList[k] = right[j]
        j = j + 1
        k = k + 1 
        
    return unSortedList   
            
for test in testCases(totalTests):
    print("input--> ",test,"\noutput--> ",mergeSort(test.copy()),"\n")
```

```python
#function for Quick sort
def partition(array,low,high,count):
    
    pivot = array[high]
    
    i = low - 1
    
    for j in range(low,high):
        count = count + 1
        if array[j] <= pivot:
            i = i + 1
            array[i],array[j] = array[j],array[i]
    
    array[high],array[i+1] = array[i+1],array[high]
        
    return i + 1,count

def quickSort(array,low,high):
    count = 0
    if low < high:
        pi,temp = partition(array,low,high,count)
        count = count + temp
        quickSort(array,low,pi-1)
        quickSort(array,pi+1,high)
    if low == 0 and high == len(array)-1:
            print("list len-->",len(array),"\niterations-->",count)
    return array

for test in testCases(totalTests):
    print("input--> ",test,"\noutput--> ",quickSort(test.copy(),0,len(test)-1),"\n")
```

```python
#function for Heap sort
def heapfiy(array,N,pos,count):
    count = count + 1
    largest = pos
    left = pos*2 + 1
    right = pos*2 + 2
    if left < N and array[left] >= array[largest] :
        largest = left  
    if right < N  and array[right] >= array[largest]:
        largest = right    
    temp = 0
    if largest != pos:
        array[largest],array[pos] = array[pos],array[largest]
        temp = heapfiy(array,N,largest,count)
    count = count + temp
    return count

def heapSort(array):
    count = 0
    for i in range(len(array)//2-1,-1,-1):  
        count = count + heapfiy(array,len(array),i,count)
    
    for i in range(len(array)-1,0,-1):
        array[0],array[i] = array[i],array[0]
        count = count + heapfiy(array,i,0,count)
    print(count)
    return array

totalTests = 10    
for test in testCases(totalTests):
    print("input--> ",test,"\noutput--> ",heapSort(test.copy()),"\n")
```
